### Тема “Множество”

#### Задание 1
Даны три множества a,b и с. Необходимо выполнить все изученные виды бинарных операций над всеми комбинациями множеств.

```python
#Решение. Зададим множества А,B и С.
A = {1, 2, 3, 4, 5, "десять", "двадцать"}
B = {4, 5, 6, 7, 8, "десять", "тридцать"}
C = {6, 2, 1, 0, 4, "один", "три"}
```

```python
#Объединение 
print(A | B | C)
```

```python
#Пересечение
print(A & B & C)
```

```python
#Разность
print(A - B)
print(A - C)
print(B - A)
print(B - C)
print(C - A)
print(C - B)
```

```python
#Симметричная разность
print(A ^ B ^ C)
```
